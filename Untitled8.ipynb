```python
# Задание task_03_01_02.

def cleared_data(data):
    """Вернуть список без значений None.

    Параметры:
        data (list): Список чисел с возможными None значениями.

    Результат:
        list: Новый список без None значений.
    """
    return [x for x in data if x is not None]

def avg(data):
    """Вернуть среднее значение 'data'.

    Параметры:
        data (list): Список чисел.

    Результат:
        float: Среднее значение элементов списка. Возвращает 0, если список пуст.
    """
    if not data:
        return 0
    return sum(data) / len(data)




n = int(input("Введите количество измерений: "))
data = []
for i in range(1, n + 1):
    measurement = input(f"Измерение {i}-е: ")
    if measurement == "-":
        data.append(None)
    else:
        data.append(int(measurement))


# Если с клавиатуры вводится прочерк, измерения нет
# (необходимо добавить в список None)

# Получить очищенный список и среднее значение
cleared_data_list = cleared_data(data)
average_temperature = avg(cleared_data_list)

print(f"\nКол-во измерений: {n}")
for i, val in enumerate(data):
  if val is None:
    print(f"Измерение {i+1}-e: -")
  else:
    print(f"Измерение {i+1}-e: {val}")

print(f"Средняя темпуратура: {average_temperature:.2f}")


# --------------
# Пример вывода:
#
# Кол-во измерений: 3
# Измерение 1-е: 10
# Измерение 2-е: -
# Измерение 3-е: 20
# Средняя температура: 15.00
```

```python
# Задание task_03_01_05.

def is_leap(year):
    """Является ли 'year' високосным годом?

    Параметры:
        year(int):Год.

    Результат:
        bool: True - да, False - нет.
    """
    return(year %4==0 and year%100!=0) or year %400==0


def days(month, year):
    """Вернуть количество дней в месяце 'month' года 'year'.

    Параметры:
        month(int): Мясяц (1-12).
        year (int): Год.

    Результат:
        int: Количество дней в месяце.
    """
    days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 21, 30, 31]
    if is_leap(year) and month ==2:
      return 29
    return days_in_month[month]


def another_date(day, month, year, delta=1):
    """Возвращает дату, смещенную на delta дней.

    Параметры:
        day(int): День.
        month(int): Месяц (1-12).
        year(int): Год.
        delta(int): Количество дней для смещения (положительное - вперед, отрицательное - назад).

    Результат:
        tuple: Кортеж (день, месяц, год) новой даты.
    """

    def previous_date_local(day, month, year):
        if day ==1:
          if month==1:
            year -=1
            month=12
            day = days(month, year)
          else:
            month-=1
            day = days(month, year)
        return day, month, year

    def next_date_local(day, month, year):
        days_in_current_month = days(month, year)
        if day == days_in_current_month:
            day = 1
            if month == 12:
                month = 1
                year += 1
            else:
                month += 1
        else:
            day += 1
        return day, month, year

    for _ in range(abs(delta)):
        if delta > 0:
            day, month, year = next_date_local(day, month, year)
        else:
            day, month, year = previous_date_local(day, month, year)

    return day, month, year

day, month, year = map(int, input("День, месяц, год через пробел: ").split())
delta = int(input("Сдвиг (может быть отрицательным): "))

new_day, new_month, new_year = another_date(day, month, year, delta=delta)

print(f"Новый день: {new_day}/{new_month}/{new_year}")



# --------------
# Пример вывода:
#
# День, месяц, год через пробел: 1 1 2000
# Свдиг (может быть отрицательным): -2
# Новый день: 30/12/1999
#
# День, месяц, год через пробел: 1 1 2000
# Свдиг (может быть отрицательным): 2
# Новый день: 03/01/2000
```

```python
# Задание task_03_01_06.
#
# Выполнил: Фамилия И.О.
# Группа: !!!


def gcd(first, second):
    while(second):
        first, second = second, first % second
    return first

def lcm(first, second):
    return (first * second) // gcd(first, second)

def gcd_nums(nums):
    result = nums[0]
    for i in range(1, len(nums)):
        result = gcd(result, nums[i])
    return result

def lcm_nums(nums):
    result = nums[0]
    for i in range(1, len(nums)):
        result = lcm(result, nums[i])
    return result

nums_str = input("Введите числа через пробел: ")
nums = [int(x) for x in nums_str.split()]

gcd_result = gcd_nums(nums)
lcm_result = lcm_nums(nums)

print(f"НОД = {gcd_result}")
print(f"НОК = {lcm_result}")

# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48
```

```python
# Задание task_03_01_11.
def ceasar(text, shift):
    """Вернуть измененную строку 'text' со сдвигом 'shift'.

    Параметры:
        - text (str): строка;
        - shift (int): свдиг.

    Результат:
        str: измененная строка."""
    letters = [chr(i) for i in range(ord('а'), ord('я') + 1)]
    result = ''
    for char in text:
        if 'а' <= char <= 'я':
            index = letters.index(char)
            new_index = (index + shift) % len(letters)  # Используем % для обработки циклического сдвига
            result += letters[new_index]
        else:
            result += char  # Если символ не буква, оставляем его без изменений
    return result

text = input("Введите предложение: ")
shift = int(input("Введите сдвиг: "))

encoded = ceasar(text, shift)
decoded = ceasar(encoded, -shift) # Для расшифровки используем обратный сдвиг (-shift)

print("Зашифрованная строка:", encoded)
print("Расшифрованная строка:", decoded)



# --------------
# Пример вывода:
#
# Введите предложение: ПрограММиРОВание С++
# Введите сдвиг: 4
# Зашифрованная строка: УфтзфдРРмФТЖдсмй Х++
# Расшифрованная строка: ПрограММиРОВание С++
```

```python
# Задание task_03_01_14.
#
# Выполнил: Фамилия И.О.
# Группа: !!!



def split_numbers(*args):
    """Вернуть 2 списка - отрицательных и неотицательных значений.

    Параметры:
        - *args: неограниченное количество числовых аргументов.

    Результат:
        tuple: кортеж из двух списков (отрицательные, неотрицательные)."""
    negative_numbers = []
    non_negative_numbers = []

    for num in args:  # Перебираем все числа, переданные в функцию
        if num < 0:   # Проверяем, является ли число отрицательным
            negative_numbers.append(num)  # Добавляем отрицательное число в список отрицательных чисел
        else:
            non_negative_numbers.append(num) # Добавляем неотрицательное число в список неотрицательных чисел

    negative_numbers.sort(reverse=True)  # Сортируем отрицательные числа по убыванию
    non_negative_numbers.sort()  # Сортируем неотрицательные числа по возрастанию

    return negative_numbers, non_negative_numbers  # Возвращаем кортеж из двух списков


print(split_numbers(1, 4, -5, 0, -33))


# --------------
# Пример вывода:
#
# ([-5, -33], [0, 1, 4])
```

```python
# Задание task_03_01_17.
#
# Выполнил: Фамилия И.О.
# Группа: !!!


def digits_sum(value):
    """Вернуть сумму цифр числа 'value'. Рекурсивная реализация.

    Параметры:
        - value: натуральное число.

    Результат:
        Сумма цифр числа.
    """
    if value < 10:
        return value
    else:
        return value % 10 + digits_sum(value // 10)


def digits_count(value):
    """Вернуть количество цифр числа 'value'. Рекурсивная реализация.

    Параметры:
        - value: натуральное число.

    Результат:
        Количество цифр в числе.
    """
    if value < 10:
        return 1
    else:
        return 1 + digits_count(value // 10)


print(digits_sum(12345))
print(digits_count(12345))



# --------------
# Пример вывода:
#
# 15
# 5
```
